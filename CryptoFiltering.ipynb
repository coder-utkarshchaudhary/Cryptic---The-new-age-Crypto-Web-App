```python
import numpy as np,pandas as pd
```

```python
df = pd.read_excel("DataExcel.xlsx")
```

```python
df
```

```python
df = df[["Date","Volume","Price","MarketCap","Solution"]]
df.nunique()
```

```python
df_lil = df.head(10000)
df_lil
```

```python
dt = df
ware = []
for unique_item in dt["Solution"].unique():
    f1 = dt[dt["Solution"]==unique_item]
    f2 = list(f1.loc[:,"Price"])
    if f2[0]==0 or f2[-1]==0: continue
    f3 = [unique_item,min(f2),max(f2),f2[-1]-f2[0],(f2[-1]-f2[0])/f2[0]*100,f2[-1]]
    ware.append(f3)
ware
```

```python
ware
```

```python
Store = pd.DataFrame(ware,columns=["Code","Min","Max","Profit","PercentChange","Current"])
Store = Store.sort_values(by=["PercentChange"],ascending=False)
Store.to_csv("DataSorted.csv",index=False)
Store
```

```python
maxDat = Store[1:6].to_numpy().T.tolist()
minDat = Store.tail().to_numpy().T.tolist()

minDat,maxDat
```

```python
import json,math
def jsonify(dat):
    dic = {"Symbol":dat[0],
           "Min":dat[1],
           "Max":dat[2],
           "Profit":dat[3],
           "PercentChange":list(np.abs(dat[4]))}
    j = json.dumps(dic,indent=4)
    print(j)
    return j

minJ = jsonify(minDat)
maxJ = jsonify(maxDat)
with open("min.json","w") as f: f.write(minJ)
with open("max.json","w") as f: f.write(maxJ)
```
